```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Read the dataset
file_path = 'Diwali Sales Data.csv'
diwali_sales_data = pd.read_csv(file_path, encoding='ISO-8859-1')

```

```python
# Data Cleaning
# Dropping unnecessary columns
diwali_sales_data_cleaned = diwali_sales_data.drop(columns=['unnamed1', 'Status'], errors='ignore')

# Filling missing values in 'Amount' column with the median
median_amount = diwali_sales_data_cleaned['Amount'].median()
diwali_sales_data_cleaned['Amount'].fillna(median_amount, inplace=True)

```

```python
# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# EDA

# 1. Distribution of key variables: Age, Gender, Occupation
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Age Distribution
sns.histplot(diwali_sales_data_cleaned['Age'], bins=30, ax=axes[0], kde=True)
axes[0].set_title('Age Distribution')

# Gender Distribution
sns.countplot(x='Gender', data=diwali_sales_data_cleaned, ax=axes[1])
axes[1].set_title('Gender Distribution')

# Occupation Distribution
sns.countplot(y='Occupation', data=diwali_sales_data_cleaned, ax=axes[2])
axes[2].set_title('Occupation Distribution')
plt.tight_layout()


```

```python
# 2. Sales trends across different product categories
plt.figure(figsize=(10, 6))
sns.countplot(y='Product_Category', data=diwali_sales_data_cleaned)
plt.title('Sales by Product Category')

```

```python
# 3. Relationship between customer attributes and spending
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Amount', hue='Gender', data=diwali_sales_data_cleaned)
plt.title('Relationship between Age and Spending by Gender')


```
